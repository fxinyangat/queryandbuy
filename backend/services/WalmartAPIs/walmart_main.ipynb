```python
def search_walmart(api_key, base_url,platform,search_query):
    url_obj = f"{base_url}platform={platform}&search={search_query}&page=1&api_key=" + API_KEY

    print("Search Complete")
    response = requests.get(url_obj)
    results = response.json()
    return results

search_walmart(
    "12499ec482f9d60e242775d3a528017985275b59",
    "https://data.unwrangle.com/api/getter/?" ,
    "walmart_search",
    "toys for kids"
    )
```

```python
def walmart_product_details(api_key, base_url,platform,product_id):
    
    url_obj = f'{base_url}platform={platform}&item_id={product_id}&api_key={API_KEY}'


    print("Product Details Complete")
    response = requests.get(url_obj)
    results = response.json()
    return results

walmart_product_details(
    "12499ec482f9d60e242775d3a528017985275b59",
    "https://data.unwrangle.com/api/getter/?" ,
    "walmart_detail",
    "16804559807")
```

```python
def walmart_product_reviews(api_key, base_url,platform,product_url,page=1):
    
    url_obj= f'{base_url}url={product_url}&page={page}&platform={platform}&api_key={API_KEY}'


    print("Product Reviews Complete")
    response = requests.get(url_obj)
    results = response.json()
    return results

walmart_product_reviews(
    "12499ec482f9d60e242775d3a528017985275b59",
    "https://data.unwrangle.com/api/getter/?",
    "amazon_reviews",
    "https://www.amazon.com/Solid-Color-Plain-Formal-Black/dp/B095PR1C6D",
    page=1)
```
